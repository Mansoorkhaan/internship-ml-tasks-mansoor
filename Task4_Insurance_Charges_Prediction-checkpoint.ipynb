# Task 4: Predicting Insurance Claim Amounts

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Load dataset
df = pd.read_csv(r'C:\Users\Administrator\.ipynb_checkpoints\insurance.csv')
print(df.shape, df.head())
```

```python
# Encode categorical variables only if they exist
df.columns = df.columns.str.lower()
categorical_cols = ['sex', 'smoker', 'region']
present_cats = [col for col in categorical_cols if col in df.columns]
if present_cats:
    df = pd.get_dummies(df, columns=present_cats, drop_first=True)
else:
    print("No expected categorical columns found for encoding.")
```

```python
# Prepare features and target
X = df.drop('charges', axis=1)
y = df['charges']
df
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Identify numeric columns to scale
num_cols = ['age', 'bmi', 'children']
intersect_cols = [col for col in num_cols if col in X_train.columns]
```

```python
# Scale numeric features
scaler = StandardScaler()
X_train[intersect_cols] = scaler.fit_transform(X_train[intersect_cols])
X_test[intersect_cols] = scaler.transform(X_test[intersect_cols])
```

```python
# Training LR model
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
# Evaluation
print("Insurance MAE:", mean_absolute_error(y_test, y_pred))
print("Insurance RMSE:", mean_squared_error(y_test, y_pred) ** 0.5)
```

```python
# Visualizations
plt.figure(figsize=(6, 4))
sns.scatterplot(x=y_test, y=y_pred)
plt.xlabel("Actual Charges")
plt.ylabel("Predicted Charges")
plt.title("Actual vs Predicted Insurance Charges")
plt.show()
sns.histplot(y_test - y_pred, kde=True)
plt.title("Distribution of Prediction Errors (Residuals)")
plt.xlabel("Residual")
plt.show()
```
