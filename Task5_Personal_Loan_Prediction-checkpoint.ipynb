# Task 5: Personal Loan Acceptance Prediction

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r'C:\Users\Administrator\.ipynb_checkpoints\bank.csv')
print(df.shape)
print(df.head())
```

```python
# Convertion of  target to binary
df['deposit'] = df['deposit'].map({'yes': 1, 'no': 0})
```

```python
# One-hot encode categorical features
df = pd.get_dummies(df, drop_first=True)
```

```python
# Defining the features and target
X = df.drop('deposit', axis=1)
y = df['deposit']
```

```python
# Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Scale numeric columns
from sklearn.preprocessing import LabelEncoder, StandardScaler
numeric_cols = ['age', 'balance', 'duration', 'campaign', 'pdays', 'previous', 'day']
scaler = StandardScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])
```

```python
# Train logistic regression and decision tree models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)
y_pred_log = log_model.predict(X_test)
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)
y_pred_tree = tree_model.predict(X_test)
```

```python
# Evaluation
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
def evaluate(model_name, y_true, y_pred):
    print(f"== {model_name} ==")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Confusion Matrix:", confusion_matrix(y_true, y_pred))
    print(classification_report(y_true, y_pred))

evaluate("Logistic Regression", y_test, y_pred_log)
evaluate("Decision Tree", y_test, y_pred_tree)
```

```python
# Decision tree feature importance
importances = pd.Series(tree_model.feature_importances_, index=X.columns).sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=importances[:10], y=importances.index[:10])
plt.title("Top 10 Influential Features (Decision Tree)")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```

```python
# Coefficients and their corresponding feature names
coef_df = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': log_model.coef_[0]
}).sort_values(by='Coefficient', ascending=False)
print("Top Positive Influencers for Loan Acceptance:")
print(coef_df.head(10))
print("\nTop Negative Influencers for Loan Acceptance:")
print(coef_df.tail(10))
```
